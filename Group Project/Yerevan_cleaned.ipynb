# **Reading data**

```python
hotel_full_Yerevan = pd.read_csv('hotel_full_Yerevan.csv')
hotel_full_Yerevan
```

```python
hotel_full_Yerevan = hotel_full_Yerevan.drop(labels = 0, axis = 0)
hotel_full_Yerevan
```

```python
nan_value = float("Nan")
hotel_full_Yerevan.replace("", nan_value, inplace = True)
hotel_full_Yerevan.dropna(inplace= True)
hotel_full_Yerevan
```

```python
hotel_full_Yerevan["city"]= hotel_full_Yerevan["city"].replace({"Երևան, Armenia" : "Yerevan, Armenia"})
```

```python
hotel_full_Yerevan["city"]= hotel_full_Yerevan["city"].replace({"Ереван, United States" : "Yerevan, Armenia"})
```

```python
hotel_full_Yerevan["city"]= hotel_full_Yerevan["city"].replace({"Երևան, Yerevan, Armenia" : "Yerevan, Armenia"})
```

```python
hotel_full_Yerevan["city"]= hotel_full_Yerevan["city"].replace({"Yerevan , Yerevan, Armenia" : "Yerevan, Armenia"})
```

```python
hotel_full_Yerevan["city"]= hotel_full_Yerevan["city"].replace({"Yerevan, Armenia, Armenia" : "Yerevan, Armenia"})
```

```python
city_list = []
for i in hotel_full_Yerevan.index:
    text = re.sub(",", "", hotel_full_Yerevan["city"][i][:hotel_full_Yerevan["city"][i].index("Armenia")])
    city_list.append(text)

hotel_full_Yerevan["city"]= city_list
hotel_full_Yerevan
```

```python
nan_value = float("Nan")
hotel_full_Yerevan.replace("", nan_value, inplace = True)
hotel_full_Yerevan.dropna(inplace= True)
hotel_full_Yerevan
```

```python
c_l = []    
for k in hotel_full_Yerevan["city"]:
    tex = re.search("\w+", k).group()
    c_l.append(tex)
hotel_full_Yerevan["city"]= c_l
hotel_full_Yerevan
```

```python
for i in hotel_full_Yerevan.index:
    hotel_full_Yerevan["country"][i] = "Armenia"
   
```

```python
hotel_full_Yerevan
```

```python
host_since_list = []
for i in hotel_full_Yerevan.host_since:
    k = re.findall("(\d{4})", i)
    host_since_list.append(k)
hs = []   
host_since_list = [item for sublist in host_since_list for item in sublist]
for k in host_since_list:
    a = int(k)
    hs.append(a)
hotel_full_Yerevan.host_since = hs 
hotel_full_Yerevan.host_since
```

```python
hotel_full_Yerevan
```

```python
string = []
for i in hotel_full_Yerevan.index:
    st = hotel_full_Yerevan.host_identity_verified[i].strip('[]').strip("'")
    string.append(st)
string
```

```python
hotel_full_Yerevan.host_identity_verified = string
hotel_full_Yerevan.host_identity_verified
```

```python
for i in hotel_full_Yerevan.index:
    if "Identity verified" in  hotel_full_Yerevan.host_identity_verified[i]:
        hotel_full_Yerevan.loc[i, 'host_identity_verified'] = "Identity verified"
    else:
        hotel_full_Yerevan.loc[i, 'host_identity_verified'] = "Identity not verified"
```

```python
hotel_full_Yerevan
```

```python
for i in hotel_full_Yerevan.index:
    if "Review" in  hotel_full_Yerevan.review[i]:
        hotel_full_Yerevan.loc[i, 'review'] = re.search("(\d+)", hotel_full_Yerevan.review[i]).group()
        
    else:
        hotel_full_Yerevan.loc[i, 'review'] = 0
    
```

```python
hotel_full_Yerevan
```

```python
for i in hotel_full_Yerevan.index:
    if "Superhost" in  hotel_full_Yerevan.superhost[i]:
        hotel_full_Yerevan.loc[i, 'superhost'] = "Super host"
    else:
        hotel_full_Yerevan.loc[i, 'superhost'] = "regular host"
```

```python
prop_list = []
for i in hotel_full_Yerevan.property_type:
    k = re.sub("hosted\sby\s\w*", "", i).strip(" ")
    prop_list.append(k)
prop_list  
```

```python
hotel_full_Yerevan.property_type =prop_list
```

```python
hotel_full_Yerevan
```

```python
accomodates_list = []
for i in hotel_full_Yerevan.accomodates:
    k = re.findall("(\d+)\sguest", i)
    accomodates_list.append(k)
number = []   
accomodates_list = [item for sublist in accomodates_list for item in sublist]
for k in accomodates_list:
    a = int(k)
    number.append(a)
    
hotel_full_Yerevan.accomodates = number 

    
```

```python
for i in hotel_full_Yerevan.index:
    if "Shared half-bath" in  hotel_full_Yerevan.number_bathroom[i]:
        hotel_full_Yerevan.loc[i, 'number_bathroom'] = 0.5
    elif "Half-bath" in  hotel_full_Yerevan.number_bathroom[i]:
        hotel_full_Yerevan.loc[i, 'number_bathroom'] = 0.5
    elif "bath" not in  hotel_full_Yerevan.number_bathroom[i]:
        hotel_full_Yerevan.loc[i, 'number_bathroom'] = 0
    elif "private bath" in  hotel_full_Yerevan.number_bathroom[i]:
        t = re.findall("(\d+(?:\.\d+)?)\sprivate", hotel_full_Yerevan.number_bedroom[i])
        hotel_full_Yerevan.loc[i, 'number_bathroom'] = t
    elif "shared bath" in  hotel_full_Yerevan.number_bathroom[i]:
        sh = re.findall("(\d+(?:\.\d+)?)\sshared\sbath", hotel_full_Yerevan.number_bedroom[i])
        hotel_full_Yerevan.loc[i, 'number_bathroom'] = sh    
    else:
        
        k = re.findall("(\d+(?:\.\d+)?)\sbath", hotel_full_Yerevan.number_bedroom[i])
        k = "".join(k)
        if k != "":
            hotel_full_Yerevan.loc[i, 'number_bathroom'] = k 
        
    
```

```python
number_beds_list = []
for i in hotel_full_Yerevan.index:
    if "bed" in  hotel_full_Yerevan.number_beds[i]:
        try:
            k = re.search("(\d+)\sbe\w{1,2}\s", hotel_full_Yerevan.number_beds[i]).group(1)
        except AttributeError:
            k = re.search("(\d+)\sbe\w{1,2}\s", hotel_full_Yerevan.number_beds[i])
    
        number_beds_list.append(k) 

number_beds_list 
```

```python
len(hotel_full_Yerevan.number_beds)
```

```python
len(number_beds_list)
```

```python
hotel_full_Yerevan.number_beds = number_beds_list
```

```python
hotel_full_Yerevan.number_beds
```

```python
for i in hotel_full_Yerevan.index:
    if "Studio" in  hotel_full_Yerevan.number_bedroom[i]:
        hotel_full_Yerevan.loc[i, 'number_bedroom'] = 0
    else:
        k = re.findall("(\d+)\sbedroom", hotel_full_Yerevan.number_bedroom[i])
        hotel_full_Yerevan.loc[i, 'number_bedroom'] = k 
        
    
```

```python
for i in hotel_full_Yerevan.index:
    if "Cleanliness" in  hotel_full_Yerevan.cleanliness[i]:
        hotel_full_Yerevan.loc[i, 'cleanliness'] = re.findall("Cleanliness\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.cleanliness[i])
    else:
        hotel_full_Yerevan.loc[i, 'cleanliness'] = None
```

```python
for i in hotel_full_Yerevan.index:
    if "Accuracy" in  hotel_full_Yerevan.accuracy[i]:
        hotel_full_Yerevan.loc[i, 'accuracy'] = re.findall("Accuracy\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.accuracy[i])
    else:
        hotel_full_Yerevan.loc[i, 'accuracy'] = None
```

```python
for i in hotel_full_Yerevan.index:
    if "Communication" in  hotel_full_Yerevan.communication[i]:
        hotel_full_Yerevan.loc[i, 'communication'] = re.findall("Communication\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.communication[i])
    else:
        hotel_full_Yerevan.loc[i, 'communication'] = None
```

```python

for i in hotel_full_Yerevan.index:
    if "Location" in  hotel_full_Yerevan.location[i]:
        try:
            hotel_full_Yerevan.loc[i, 'location'] = re.search("Location\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.location[i]).group(1)
        
        except AttributeError:
            hotel_full_Yerevan.loc[i, 'location'] = re.search("Location\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.location[i])
    
    else:
        hotel_full_Yerevan.loc[i, 'location'] = None

 
```

```python
for i in hotel_full_Yerevan.index:
    if "Check-in" in  hotel_full_Yerevan.check_in[i]:
        hotel_full_Yerevan.loc[i, 'check_in'] = re.findall("Check-in\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.check_in[i])
    else:
        hotel_full_Yerevan.loc[i, 'check_in'] = None
```

```python
for i in hotel_full_Yerevan.index:
    if "Value" in  hotel_full_Yerevan.value[i]:
        hotel_full_Yerevan.loc[i, 'value'] = re.findall("Value\s(\d+(?:\.\d+)?)", hotel_full_Yerevan.value[i])
    else:
        hotel_full_Yerevan.loc[i, 'value'] = None
```

```python
for i in hotel_full_Yerevan.index:
    if "No reviews (yet)" in  hotel_full_Yerevan.rating[i]:
        hotel_full_Yerevan.loc[i, 'rating'] = 0
    
```

```python
price_list = []
for i in hotel_full_Yerevan.price:
    k = re.findall("(\d+)\n", i)
    price_list.append(k)

price_list = [item for sublist in price_list for item in sublist]
  
hotel_full_Yerevan.price = price_list
```

```python
for i in hotel_full_Yerevan.index:
    if "Response rate" in  hotel_full_Yerevan.host_response_rate[i]:
        t = re.findall("\s(\d+)", hotel_full_Yerevan.host_response_rate[i])
        hotel_full_Yerevan.loc[i, 'host_response_rate'] = t 
    else:
        hotel_full_Yerevan.loc[i, 'host_response_rate'] = None
        
```

```python
convert_dict = {'host_since': float, 'host_response_rate': float,  'review': float,  'accomodates': float, 
                                        'number_bathroom': float, 'number_bedroom': float, 
                                        'number_beds': float,'rating': float,  'price': float,
                                        'cleanliness': float, 'accuracy': float, 'communication': float, 
                           'location': float, 'check_in': float, 'value': float}
```

```python
hotel_full_Yerevan = hotel_full_Yerevan.astype(convert_dict)
```
