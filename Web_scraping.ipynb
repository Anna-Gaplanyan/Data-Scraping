```python
Data_first_page = pd.DataFrame({'link':[''], 'Code':[''], 'Sale_rent':[''], 'Price': [''], 'Total_area':[''], 
                      'Floor':[''], 'Rooms':[''], 'Date':['']})
url = 'https://akcern.am/en/category?city=&property-type=72&realty-type=&page=1&per-page=15'
page = requests.get(url)
soup_all = BeautifulSoup(page.text, 'lxml')
postings = soup_all.find_all('div', class_ = 'col-xl-4 col-lg-6 col-md-4 col-sm-6 col-12')
for post in postings:
    link = base + post.find('a', {'class': "property-img"}).get('href')
    code = post.find('div', class_ = "card-code").text.strip()
    sale_rent = post.find('h1', class_ = "title").text.strip()
    price = post.find('p', class_ = "price").text
    feature = post.find("ul", class_ = 'facilities-list clearfix').text.strip()
    total_area=str(re.findall("\w\w\w\w\w\s\w\w\w\w\s\d*\s\w\s\d*", feature)).strip('[]')
    floor = str(re.findall("F\w\w\w\w\s\d*", feature)).strip('[]')
    room = str(re.findall("C\w\w\w\w\s\w\w\s\w\w\w\w\w\s\d*", feature)).strip('[]')
    date = post.find('div', class_ = 'footer').text.strip()
    Data_first_page = Data_first_page.append({'link':link, 'Code':code, 'Sale_rent':sale_rent, 'Price': price, 'Total_area':total_area, 
                        'Floor':floor, 'Rooms':room, 'Date':date}, ignore_index= True)
```

# **Question 2**

```python
location_list = []

for i in link_list:
    page_requests = requests.get(i)
    page_requests_soup = BeautifulSoup(page_requests.text, 'lxml')
    location = page_requests_soup.find('div', class_ = "detail-location").text.strip()
    location_list.append(location)
    
location_list
    

```

```python
Data['Location'] = location_list
```

```python
Data
```

```python
pages_list = []

for i in link_list:
    page_requests_feat = requests.get(i)
    page_resp = TextResponse(body = page_requests_feat.text, url = page_requests_feat.url, encoding = 'utf-8')
    page_xpath = page_resp.xpath("//span[@class = 'font-weight-bold']/text()")
    pages_list.append(page_xpath)
    
pages_list = pages_list = [item for sublist in pages_list for item in sublist]
pages_list_extract = []
for p in pages_list:
    page_e = p.extract()
    pages_list_extract.append(page_e)
    
pages_list_extracted = []
for ex in pages_list_extract:
    ext = ex.strip()
    pages_list_extracted.append(ext)
pages_list_extracted
```

```python
pages_list_yn = []

for i in link_list:
    page_requests_yn = requests.get(i)
    page_resp_yn = TextResponse(body = page_requests_yn.text, url = page_requests_yn.url, encoding = 'utf-8')
    page_xpath_yn = page_resp_yn.xpath("//img[@class = 'status-icon']")
    pages_list_yn.append(page_xpath_yn)
    
pages_list_yn =[item for sublist in pages_list_yn for item in sublist]
image_list = []
for z in pages_list_yn:
    alt = z.extract()
    image_list.append(alt)

image_list_yn = []
for t in range(len(image_list)):
    alt = re.findall("alt\=\"(\w*)", image_list[t])
    image_list_yn.append(alt)

image_list_yn = [item for sublist in image_list_yn for item in sublist]
image_list_yn
```

```python
full_feature = []
for (item1, item2) in zip(pages_list_extracted, image_list_yn):
    full_feature.append(item1 + item2)
        

full_feature   
```

```python
string = ','.join(full_feature)
feature_list = re.findall("Air\sConditioner\(AC\)\:\w*\,Garage\:\w*\,Parking\:\w*\,Furniture\:\w*\,Appliances\:\w*\,Balcony\:\w*\,Condition\:\w*", string)
feature_list
```

```python
Data['Features']=feature_list
Data['Features']
```

```python
from datetime import date
today = date.today()
today = today.strftime("%b-%d-%Y")


Data.to_csv('Akcern_first_page' + today + '.csv', index = False)

```
